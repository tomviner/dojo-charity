```python
%matplotlib notebook
import pandas as pd
```

```python
df = pd.read_csv('charities.csv')

pd.options.display.float_format = '£{:,.2f}'.format
```

```python
df.keys()
```

```python
df.dropna(subset=['income', 'spending'], inplace=True)
df = df[(df['spending'] != 0) & (df['income'] != 0)]
```

```python
df
```

```python
df['retained'] = df['income'] - df['spending']
df['spending_ratio'] = df['retained'] / df['spending']
```

```python
df[['retained', 'spending_ratio']]
```

```python
df.plot.scatter(x='income', y='spending')
```

```python
df.dropna(subset=['spending_ratio'], inplace=True)
```

```python
df = df[df['spending_ratio'] != 0]

df['spending_ratio'] = df['spending_ratio'].map('{:,.2f}'.format)
```

```python
df.sort_values('retained', inplace=True)
```
